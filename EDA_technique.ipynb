# importing libraies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Read the data

```python
df=pd.read_csv('breast_cancer_data.csv.csv')
```

showing first 5 records

```python
df.head(5)
```

information of the table using** info method **

```python
df.info()
```

```python
df=df.drop  (['Unnamed: 32','id'],axis=1)
```

describe the columns only numeric columns

```python
df.describe()
```

identifying missing values

```python
df.isnull().sum()
```

duplicate values

```python
df.duplicated().sum()
```

map B:0 and M :1

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
```

distribution target variable

```python
df['diagnosis'].value_counts()
```

```python
(df['diagnosis'].value_counts() / len(df) * 100).round(2).astype(str) + '%'
```

Visualize target distribution

```python
sns.countplot(x='diagnosis', data=df)
plt.show()
```

Histogram for all numeric features

```python
df.hist(figsize=(15,15))
plt.show()
```

outliers using boxplot

```python
plt.figure(figsize=(30,15))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

pairplot on dataset few columns

```python
sns.pairplot(df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'diagnosis']], hue='diagnosis')
plt.show()
```

Distribution plots for numeric features

```python

numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
numeric_cols.remove('diagnosis')

plt.figure(figsize=(20,15))
for i, col in enumerate(numeric_cols):
    plt.subplot(6,5,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

corelation between columns using heatmp

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=df)
plt.show()
```
